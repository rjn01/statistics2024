### Problem 1(a)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
h = np.random.uniform(low=140, high=200, size=20)
g = np.random.choice(['female', 'male'], 20, replace=True, p=[0.6, 0.4])
df = pd.DataFrame({'height': h, 'gender': g})
print(df) 
```

### Problem 1(b)

```python
df.at[0, 'height'] = 120  # Replacing height of first entry with 120.
```

```python
df.loc[0] # checking data of first entry
```

```python
plt.figure(figsize=(8, 6))
df.boxplot(by='gender', column='height', grid=False)
plt.title('Height Distribution by Gender (Pandas Boxplot)')
plt.suptitle('')  # Removes the automatic Pandas boxplot title
plt.xlabel('Gender')
plt.ylabel('Height')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='gender', y='height')
plt.title('Height Distribution by Gender (Seaborn Boxplot)')
plt.xlabel('Gender')
plt.ylabel('Height')
plt.show()
```

Interpretations for Box Plot:
1. The median for females is lower than for males, which indicates that females tend to be shorter overall.
2. Boxes for male and female overlap significantly, there might be little difference between the height distributions of males and females.
3. The height, which is manually adjusted to 120 for females appears as an outlier, projecting it is much lower than typical values for this gender.

### Problem 1(c)

```python
female_heights = df[df['gender'] == 'female']['height']
male_heights = df[df['gender'] == 'male']['height']

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sm.qqplot(female_heights, line='q', ax=plt.gca())
plt.title('Quantile-Quantile(Q-Q) Plot - Female Heights')

plt.subplot(1, 2, 2)
sm.qqplot(male_heights, line='q', ax=plt.gca())
plt.title('Quantile-Quantile(Q-Q) Plot - Male Heights')

plt.tight_layout()
plt.show()
```

Interpretations for Q-Q Plot:
1. Points are not very close to the line so Datas are not appoximatelt normal.
2. Due to small sample it is hard to interprate Q-Q plots.
