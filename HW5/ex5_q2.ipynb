```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
txt_file = pd.read_csv('mtcars.txt', sep='\t')
xlsx_file = pd.read_excel('mtcars.xlsx')

if not txt_file.equals(xlsx_file):
    print("Files not equal!")

cars = txt_file
```

b)

```python
cars.plot.scatter(x='wt', y='mpg')
plt.xlim(0, 6)
plt.ylim(10, 40)
plt.title('wt vs mpg')
plt.xlabel('Weight')
plt.ylabel('Miles per Gallon')
plt.show()
```

c)

```python
sns.pairplot(cars)
plt.suptitle('Scatter Plot for each pair of Variables')
plt.show()
```

d) cars[’cyl’].value counts() returns the # of times each unique value occurs in the 'cyl' column

```python
cyl_vals = cars['cyl'].value_counts()
print(cyl_vals)
```

```python
cyl_vals.plot(kind='pie')
plt.title('Cylinders in Cars')
plt.show()
```

e)

```python
corr_mat = cars.corr(numeric_only=True)
sns.heatmap(corr_mat)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
